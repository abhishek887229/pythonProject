# BAR PLOT

## below is the code showing how to develop a box plot in matplotlib

explainging code
* color -->create a list of color and pass to it, it will apply different color accordingly
* align--> it align the label at x (either center or edge)
* aplha--. chnage the opaciy of grpah (0,1)
* lable--> which value you want to show in legend
* to print lables we use other object that is `plt.legend()`

```python
x=["male","female","other"]
y=[85,75,24]
plt.xlabel("gender -->",fontsize=10)
plt.ylabel("count -->")
c=['k','b','g']
plt.bar(x,y,width=0.3,color=c,align='center',animated=True,edgecolor=None,alpha=0.8,mouseover=True,label=x)
plt.legend()

plt.show()
```

## creating double Bar graph in matplotlib

let's understnad a scenario we have slaes for different year we want to show sales of two years in same bar chart we can do this in matplotlib with little trick

```python
x=[2012,2013,2014,2015,2016]
y=[500,700,900,150,985]
z=[600,400,1000,854,478]
```

```python
# set width first that will help in orgnization of bar
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
print(p)
print(p1)
# p is width for y year p1 for z
```

```python
#this figure change the size of figure or plot
plt.figure(figsize=(10,20))
plt.xlabel("year")
plt.ylabel("val")
plt.bar(p,y,width,label="sales")
plt.bar(p1,z,width,label="profit")
#xticks is responsible for label settings and allignemnt
# this p+widh/2 --> center align the object
# rotation will rotate label value
plt.xticks(p+width/2,x,rotation=90)
plt.legend()
plt.show()
```

## replace  `bar` with `barh` the axis of bar will chnage

# SCATTER PLOT

LET WE HAVE AGE AND HEIGHT OF PEOPLE SO WE HAVE DATA IN TWO LIST AND WE WILL USE THAT TO WORK WITH MATPLOTLIB AND DRAW SCATTER PLOT

```python
#AGE
age=[1,5,10,15,20,95,75,4,21]
#height in cm
h=[40,65,140,170,180,179,160,60,190]
#----------------------

plt.scatter(age,h,s=h)
plt.title("Age vs height ")
plt.xlabel("Age -->")
plt.ylabel("height  -->")
plt.colorbar()
plt.legend()
plt.show()
```

## here is c map you can use in scatter plotpng234.png

```python


# Assuming 'age', 'h', and 'c' are your data arrays
# 'age' and 'h' are arrays representing age and height
# 'c' is the array representing the values to map to the colormap

#this label is helpful to extract color from cmap so we uing
c = [10, 20, 30, 40, 50, 60, 70, 80,100]  # Dummy data for the colormap

plt.scatter(age, h, s=100, cmap='PuBu', c=c, alpha=0.7)  # Adjust 's' according to your preference
plt.xlabel("Age (in years)")
plt.ylabel("Height (in cm)")
plt.grid(True)
plt.colorbar(label="Colorbar Label")  # Add a label for the colorbar
plt.legend()  # If you have a legend, make sure it's properly defined
plt.show()
```

# Hist-plot

it use to show the distribution of data and helpful to understand nature of data
* we will use iris data from seaborn
* this will help to work with data

```python
x=sns.load_dataset('iris')
```

```python
x.head()
```

#hist creation

```python
l=[0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90]
plt.hist(x.sepal_length,color='black',edgecolor='white')
plt.xlabel("sepal_length")
plt.ylabel('numbers')

plt.show()
```

## please note that in above chart data is not arranged see the x-axis it's hard to read the distribution of bars, so we will use bins to arrange them

* if you write bins=5(data will arrange in  5 cols only)
* if write bins=20(data will arrage in 20 bins only)

```python
#l will define bins
l=[0.10,0.20,0.30,0.40,0.50,0.60,0.70,0.80,0.90]
plt.hist(x.sepal_length,color='black',edgecolor='white',bins=50)
plt.xlabel("sepal_length")
plt.ylabel('numbers')

plt.show()
```

#range in hist, cumulative parameter (it arrange hist reverse)

```python
plt.hist(x.sepal_length,"auto",color='black',edgecolor='white',cumulative=-1,bottom=5,align='mid')
plt.xlabel("sepal_length")
plt.ylabel('numbers')

plt.show()
```

# Pie Plot

---
## parameters explain
1) labels --> assign name to each segment of pi chart (labels="col name")

2) explode --> it seprates a portion of chart so it can be visualize seprate you have to pass a list of values that show how close you want to center the portions of charts are.

3) color= you can give different color to the pie chart

4) autopct --> auto percent to show percent value e.g. (%0.2f%%)

5) shadow :- give 3d effect of shadow

6) radius :- to increase radius of chart use radius parameters

7) label distance --> distance of labels

```python
y=x.groupby(x['species']).sum()
```

```python
plt.pie(y.petal_width,labels=y.index,explode=[0.4,0,0],autopct="%0.2f%%",shadow=True)
plt.show()
```

# Adding some other features


## we have some additional features in matplotlib have a look

axhline--> make a horizontal line along the axis

axvline --> make a verticle line along the axis

text --> use to write text over the grpah

plt.text(x_Position,y_position, text )

```python
plt.hist(x.sepal_length, bins=20, color='skyblue', edgecolor='black')
plt.axvline(x.sepal_length.mean(),c='r',label='mean')
plt.text(x.sepal_length.mean()+1.2,10,f"((this text is by using text function) mean is {np.round(x.sepal_length.mean(),2)}")
plt.legend()
plt.grid(True)
plt.show()
```
